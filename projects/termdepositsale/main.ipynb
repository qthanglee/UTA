# Univariate, Missing Values and Outliers

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# load data and get some insight
pdata = pd.read_csv("bank-full.csv")
pdata.head()
```

```python
unique_values = pdata.nunique()
print('Count unique values in each column')
print(unique_values)
```

```python
pdata.describe()
```

```python
# check for null value
pdata.isnull().sum()
```

```python
pdata.shape
```

```python
# Categorical Data (ordinal, nominal, binary)
# job, marital, default, education, housing, loan, contact, poutcome
```

```python
# Summarizing education category variable 
sns.countplot(x='education', data=pdata)
plt.title('Number of people by education')
plt.ylabel('Number of people')
plt.xlabel('Education')
plt.xticks(rotation = 45)
plt.show();
```

```python
# Summarizing job category variable 
sns.countplot(x='job', data=pdata)
plt.title('Number of people by job type')
plt.ylabel('Number of people')
plt.xlabel('Job')
plt.xticks(rotation = 90)
plt.show();
```

```python
# Summarizing marital category variable 
sns.countplot(x='marital', data=pdata)
plt.title('Number of people by marital')
plt.ylabel('Number of people')
plt.xlabel('Marital')
plt.xticks(rotation = 90)
plt.show();
```

```python
# Summarizing housing category variable 
sns.countplot(x='housing', data=pdata)
plt.title('Number of people by housing')
plt.ylabel('Number of people')
plt.xlabel('Housing')
plt.xticks(rotation = 90)
plt.show();
```
